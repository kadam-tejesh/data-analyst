```python
#importing dataset of netflix movies and series by pandas
import pandas as pd
df=pd.read_csv('netflix_titles.csv.zip')
df.head()
```

```python
#calculating no of null values in each column
df.isnull().sum()
```

```python
#droping duplicate values in data set
df.drop_duplicates()
df.shape
```

```python
#filling the null values 
a=df.ffill()
a.head()
```

```python
b=a.ffill(axis=1)
b.isnull().sum()
```

```python
#changing the date format 
b['date_added']=pd.to_datetime(b['date_added'],format='mixed')
b
```

```python
b['rating'].unique()
```

```python
#replacing the rating column
c=b.replace(to_replace=['PG-13', '74 min','PG', 'TV-14'],value='2')
d=c.replace(to_replace=['TV-PG', 'TV-Y', 'TV-Y7', 'R',],value='3')
e=d.replace(to_replace=['84 min', '66 min', 'NR'],value='4')
f=e.replace(to_replace=['TV-Y7-FV', 'UR','TV-MA',],value='1')




```

```python
#final clean and preprocessed dataset
dataset=f.sort_index()
dataset
```
